# CityBikes

```python
import requests
import pandas as pd
from pprint import pprint
```

### Send a request to CityBikes for the city of your choice. 

Approach - using the `/network_id` endpoint for London directly

```python
# url for endpoint for London network
url = 'https://api.citybik.es/v2/networks/santander-cycles'
```

```python
response = requests.get(url)
response
```

```python
bike_data = response.json()
# print(pprint(bike_data))
```

### Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). 

```python
# To get the details of each station in London city
london_stations = pd.json_normalize(bike_data['network']['stations'])
london_stations
```

### Put your parsed results into a DataFrame.

```python
# to get the name, longitude, latitude and number of free bikes per station in 
city_bikes = london_stations[['extra.name', 'longitude', 'latitude', 'free_bikes', 'empty_slots', 'extra.uid']]
city_bikes
```

```python
# Save city bikes dataframe
city_bikes.to_pickle("city_bikes.pkl")
```
